#Camera Filtering

```python
df['camera'].value_counts()
```

```python
a='50 MP + 8 MP + 2 MP Triple Rear & 16 MP Front Camera'
a.split()[0:2]
" ".join(a.split()[0:2])
```

```python
df['Rear Camera']=df['camera'].apply(lambda a:" ".join(a.split()[0:2]))
```

```python
a='50 MP + 8 MP + 2 MP Triple Rear & 16 MP Front Camera'
a.split()[-4:-2]
" ".join(a.split()[-4:-2])
```

```python
df['Front Camera']=df['camera'].apply(lambda a:" ".join(a.split()[-4:-2]))
```

```python
def cam(text):
    # Check for specific camera types
    if 'Dual Rear' in text:
        return "Dual Rear"
    elif 'Triple Rear' in text:
        return "Triple Rear"
    elif 'Quad Rear' in text:
        return "Quad Rear"
    # Check for 'Rear' without specific types, classify as "Other Camera Configuration"
    elif 'Rear' in text:
        return "Rear"
    # Default classification for non-camera-related descriptions
    else:
        return "Unknown Camera Type"
```

```python
df['Camera Type']=df['camera'].apply(cam)
```
