```python
def cam(text):
    # Check for specific camera types
    if 'Dual Rear' in text:
        return "Dual Rear"
    elif 'Triple Rear' in text:
        return "Triple Rear"
    elif 'Quad Rear' in text:
        return "Quad Rear"
    # Check for 'Rear' without specific types, classify as "Other Camera Configuration"
    elif 'Rear' in text:
        return "Rear"
    # Default classification for non-camera-related descriptions
    else:
        return "Unknown Camera Type"
```

```python
df['Camera Type']=df['camera'].apply(cam)
```
